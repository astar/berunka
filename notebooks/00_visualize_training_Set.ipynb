```python
#Random 250 spectra from dataset
for i, j in enumerate(np.random.permutation(X.shape[0])[:25]):
    pl.subplot(5, 5, (i + 1))
    pl.plot(X[j],c='k')
    pl.title("idx: %d, Class: %d" % (j, y[j]) )
    pl.xticks(()), pl.yticks(())
    pl.tight_layout()
[pl.savefig('pic1.{}'.format(ext)) for ext in ['png','svg','pdf']]
```
