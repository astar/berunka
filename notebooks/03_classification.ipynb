# Testing different classification algoritms and data sets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from pandas.tools.plotting import andrews_curves
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
from sklearn.cross_validation import ShuffleSplit
from sklearn import tree
pd.set_printoptions(max_rows=200, max_columns=1000)
```

```python
%pylab inline
import pylab as pl
import numpy as np

# Some nice default configuration for plots
pl.rcParams['figure.figsize'] = 10, 10
pl.rcParams['axes.grid'] = True
```

## Data preprocessing

```python
data = np.genfromtxt('data/archive_1_4_28072013.csv' , delimiter=',') # load 100A around Halpha
```

```python
#data = np.genfromtxt('data/1_5_whole_spectrum.csv' , delimiter=',') # load whole spectrum
```

```python
#data = np.genfromtxt('data/1_5_skewed_whole_spectrum.csv' , delimiter=',') # load whole spectrum
```

```python
X, y = data[:,1:], data[:,0]
print("data shape: %r, target shape: %r" % (X.shape, y.shape))
print("classes: %r" % list(np.unique(y)))
```

```python
n_samples, n_features = X.shape
print("n_samples=%d" % n_samples)
print("n_features=%d" % n_features)
```

```python
#X = preprocessing.scale(X)
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1)

print("train data shape: %r, train target shape: %r"
      % (X_train.shape, y_train.shape))
print("test data shape: %r, test target shape: %r"
      % (X_test.shape, y_test.shape))
```

## SVM

```python
clf = SVC(kernel="rbf", C=100, gamma=0.01).fit(X_train, y_train)
predicted = clf.predict(X_test)
clf.score(X_test, y_test) 
```

## Cross-validation

```python
from sklearn.cross_validation import cross_val_score

svc = SVC(kernel="rbf", C=100, gamma=0.01)
cv = ShuffleSplit(n_samples, n_iter=10, test_size=0.1,
    random_state=0)

test_scores = cross_val_score(svc, X, y, cv=cv, n_jobs=2)
test_scores
```

```python
from scipy.stats import sem

def mean_score(scores):
    """Print the empirical mean score and standard error of the mean."""
    return ("Mean score: {0:.3f} (+/-{1:.3f})").format(
        np.mean(scores), sem(scores))
```

```python
print(mean_score(test_scores))
```

```python
from sklearn import metrics
print metrics.classification_report(y_test, predicted)
```

## DT

```python
clf = tree.DecisionTreeClassifier().fit(X_train, y_train)
predicted = clf.predict(X_test)
clf.score(X_test, y_test) 
```

```python
from sklearn.cross_validation import cross_val_score

svc = tree.DecisionTreeClassifier().fit(X_train, y_train)
cv = ShuffleSplit(n_samples, n_iter=10, test_size=0.1,
    random_state=0)

test_scores = cross_val_score(svc, X, y, cv=cv, n_jobs=2)
test_scores
```

```python
print(mean_score(test_scores))
```
