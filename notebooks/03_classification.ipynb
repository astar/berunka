# Testing different classification algoritms and data sets

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from pandas.tools.plotting import andrews_curves
from sklearn.grid_search import GridSearchCV
from sklearn.svm import SVC
from sklearn.cross_validation import ShuffleSplit
from sklearn import tree
pd.set_printoptions(max_rows=200, max_columns=1000)
```

```python
%pylab inline
import pylab as pl
import numpy as np

# Some nice default configuration for plots
pl.rcParams['figure.figsize'] = 10, 10
pl.rcParams['axes.grid'] = True
```

## Data preprocessing

```python
data = np.genfromtxt('data/1_5_100.csv' , delimiter=',') # load 100A around Halpha
```

```python
data = np.genfromtxt('data/1_5_whole_spectrum.csv' , delimiter=',') # load whole spectrum
```

```python
data = np.genfromtxt('data/1_5_skewed_whole_spectrum.csv' , delimiter=',') # load whole spectrum
```

```python
X, y = data[:,1:], data[:,0]
print("data shape: %r, target shape: %r" % (X.shape, y.shape))
print("classes: %r" % list(np.unique(y)))
```

```python
#X = preprocessing.scale(X)
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1)

print("train data shape: %r, train target shape: %r"
      % (X_train.shape, y_train.shape))
print("test data shape: %r, test target shape: %r"
      % (X_test.shape, y_test.shape))
```

## SVM

```python
clf = SVC(kernel="rbf", C=1, gamma=0.001).fit(X_train, y_train)
predicted = clf.predict(X_test)
clf.score(X_test, y_test) 
```

## DT

```python
clf = tree.DecisionTreeClassifier().fit(X_train, y_train)
predicted = clf.predict(X_test)
clf.score(X_test, y_test) 
```

## Plot prediction

```python
for i, j in enumerate(np.random.permutation(X_test.shape[0])[:25]):
    pl.subplot(5, 5, (i + 1))
    pl.plot(X_test[j])
    pl.title("%d,True: %d Pred: %d" % (j, y_test[j], predicted[j]) )
    pl.xticks(()), pl.yticks(())
```

## Plot missclasification

```python
X_miss = X_test[y_test!=predicted]
y_miss = y_test[y_test!=predicted]
predicted_miss = predicted[y_test!=predicted]
```
